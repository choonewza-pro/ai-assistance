```python
import requests

url_lextop = "https://api.aiforthai.in.th/lextoplus"
url_tlexp = "https://api.aiforthai.in.th/tlexplus"
headers = {"Apikey":"S26cDzNRhoxp2xMT74DGxleS5uMrvlXT"}
```

```python
def ai4thai_word_tokenize(url, header, text):
    params = {"text":text, "norm" : "1"}
    res = requests.get(url, params=params, headers=headers).json()
    return res['tokens']
```

```python
text = "ปัญญาประดิษฐ์เป็นเทคโนโลยีที่มีบทบาทสำคัญในยุคปัจจุบันโดยเฉพาะในด้านการประมวลผลข้อมูลและการเรียนรู้ของเครื่องซึ่งสามารถนำมาใช้ในหลากหลายอุตสาหกรรม"
```
